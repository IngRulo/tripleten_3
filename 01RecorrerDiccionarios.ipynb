```python
# Ejemplo recorrido por valor
financial_info = {
    'American Express': 93.23,
    'Boeing': 178.44,
    'Coca-Cola': 45.15,
    'Walt Disney': 119.34,
    'Nike': 97.99,
    'JPMorgan':96.27,
    'Walmart': 130.68 
}

for value in financial_info.values():
    print(value)
```

```python
# Ejercicio 1
# Calcule el valor total de todas las acciones. 

financial_info = {
    'American Express': 93,
    'Boeing': 178,
    'Coca-Cola': 45,
    'Walt Disney': 119,
    'Nike': 97,
    'JPMorgan': 96,
    'Walmart': 130
}

total_value = 0 # el valor que actualizarás en cada iteración del bucle

# escribe tu bucle for aquí. En cada iteración del ciclo, extrae el precio
for s in financial_info.values():
    total_value += s
# y actualiza la variable total_value

print(total_value)
```

```python
# Ejercicio 2
# Calcular el valor total con items
financial_info = {
    'American Express': 93,
    'Boeing': 178,
    'Coca-Cola': 45,
    'Walt Disney': 119,
    'Nike': 97,
    'JPMorgan': 96,
    'Walmart': 130
}

total_value = 0

for clave, valor in financial_info.items():
    total_value += valor

print(total_value)
```
